```python
# Import libraries
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt
import glob
import wandb
import os
import datetime
```

```python
# Load model 
model_version = 'yolo11m-seg.pt' # can pick from yolov8n.pt, yolov8s.pt, yolov8m.pt, yolov8l.pt, yolov8x.pt
model = YOLO(model_version)
```

**Check model benchmarks [here](https://docs.ultralytics.com/tasks/segment/#models)**

#### Model training

```python
proj_name = 'corm_experiments'
# run = wandb.init(project=proj_name)
```

```python
# Train the model
dat = "../data/feb-25.yaml"
batch = 8
opt='auto'
exp_name = "exp_"+os.path.basename(dat).split(".")[0]+"_"+model_version.split(".")[0]+"_"+ datetime.datetime.now().strftime("%Y-%m-%d_%H_%M")

results = model.train(data=dat, epochs=200, imgsz=640, project=proj_name, patience=35, name = exp_name, 
                      batch=batch, optimizer=opt)
```

```python
# Validate the model
metrics = model.val()
```

```python
# Export model
# form = "tflite"
# model.export(format=form)
```

```python
# Make Predictions
test_data = glob.glob('../../Image-data/new-data/YOLODataset_seg/images/test/*.jpg')
results = model.predict(test_data, show_labels=True, show_conf=True, conf=0.2)
```
